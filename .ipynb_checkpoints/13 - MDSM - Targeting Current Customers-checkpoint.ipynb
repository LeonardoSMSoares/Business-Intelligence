# Targeting Current Customers

Mass marketing treats all customers as one group. Oneto-one marketing focuses on one customer at a time. Target marketing to selected groups of customers or market segments lies between mass marketing and oneto-one marketing. Target marketing involves directing marketing activities to those customers who are most likely to buy.

Targeting implies selection. Some customers are identified as more valuable than others and these more highly valued customers are given special attention. By becoming skilled at targeting, a company can improve its profitability, increasing revenues and decreasing costs. Targeting is best executed by companies that keep detailed records for individuals. These are companies
that offer loyalty programs or use a customer relationship management system. Sales transactions for individual customers need to be associated with the specific customer and stored in a customer database.

## Data Mining 

```python
# Show NA's
sns.heatmap(bank.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
bank.info()
```

```python
# Get all the categorical features of the DataFrame (without 'customerID')
cat_feats = ['job','marital','education','default','housing', 'loan','contact','month','poutcome']

# Convert "response" into a dummy variable
df = pd.get_dummies(bank,columns = ['response'], drop_first=True)

df.head()
```

```python
# Convert the categories into integers

from category_encoders import OrdinalEncoder

ordinal_encoder = OrdinalEncoder(cols=cat_feats, handle_missing='return_nan')

df = ordinal_encoder.fit_transform(df)

df.head()
```

### Split the Data

```python
X = df.drop(['response_yes'], axis=1)  # Indepedent variable
y = df['response_yes']                 # Dependent variable

y.unique()    # Verify output
```

```python
# Lets see how imbalance is this dataset looking for y=1:

sum(y) / len(y)
```

```python
# Split the data into a training and testing set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, stratify=y)
```

```python
# Let's see if stratify worked as expected, creating groups with the same proportion of 1's and 0's.

print(sum(y_train) / len(y_train))
print(sum(y_test) / len(y_test))
```

## Build LGBM Model

```python
from lightgbm import LGBMClassifier

# Create an instance of LGBMClassifier
model = LGBMClassifier()             

# Fit the data to the training set
model.fit(X_train,     
          y_train,     
          verbose=True,                 # Verbosity of printing messages
          eval_metric='aucpr',          # Area under the PR curve
          eval_set=[(X_test,y_test)])   # Train with the train dataset, but evaluete the how many trees with the test dataset
```

```python
# Make predictions
predictions = model.predict(X_test)
```

## Results Obtained

```python
# Classification metrics
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score

# Check the model's accuracy
accuracy_score(predictions, y_test)
```

```python
# Print the confusion Matrix
confusion_matrix(predictions, y_test)
```

```python
# Plot the Confussion Matrix

from sklearn.metrics import plot_confusion_matrix

sns.set_style("dark")  # take out the grid

plot_confusion_matrix(model,              # estimator
                      X_test,             # X
                      y_test,             # y-true
                      values_format='d',
                      display_labels=['Not sensible at promotions', 'responses to promotions'])
```

```python
# Print the classification report

print(classification_report(predictions, y_test))
```
